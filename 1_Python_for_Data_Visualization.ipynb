### Renaming Columns in Pandas
*two popular ways to rename columns in pandas*

**dictionary substitution** - useful only if you want to rename a few columns

**list replacement** - requires full list of names, can be more error prone


```
carFinanceCSVDataFrame.head()
```

### Data Imputation

```
# fillna will fill in NaN values uing backfill, good for time series data
carFinanceCSVDataFrame['interest_paid'][30:40].fillna(method='bfill')
```

```
# fillna will fill in NaN values uing frontfill, good for time series data
carFinanceCSVDataFrame['interest_paid'][30:40].fillna(method='ffill')
```

### Linear Interpolation

Linear Interpolation is the filling in of values

```
carFinanceCSVDataFrame['interest_paid'][30:40].interpolate(method='linear')
```

```
# interest paid before filling in the NaN with a value
carFinanceCSVDataFrame['interest_paid'].sum()
```

```
# fill in with the actual value
interest_missing = carFinanceCSVDataFrame['interest_paid'].isna()
carFinanceCSVDataFrame.loc[interest_missing, 'interest_paid'] = 93.24
```

## Matplotlib
Matplotlib is a Python library that produces quality plots, graphs and charts with 2d and 3d dimension control.

```
import pandas as pd
import numpy as np

# plt is common alias for matplotlib.pyplot module
import matplotlib.pyplot as plt

# seaborn libary is wrapper of matplotlib
import seaborn as sns

# inline flag will use backend to makes figures inline with notebook
%matplotlib inline
```

```
df = pd.read_csv(csvExportTablePath)
```

```
df.head()
```

```
df.info()
```

```
# creating seperate variables that select all rows and single column values
month_number = df.loc[:, 'month'].values
interest_paid = df.loc[:, 'interest_paid'].values
principal_paid = df.loc[:, 'principal_paid'].values
```

```
# the values attribute converts column of values into numpy array
type(month_number)
```

### Setting Marker Types and Colors for Matplotlib

```
# plotting using different markers
plt.plot(month_number, interest_paid, marker='.', markersize=10)
plt.plot(month_number, principal_paid, marker='.', markersize=10)
```

```
# plotting using different markers and colors
# https://matplotlib.org/3.1.1/tutorials/colors/colors.html
plt.plot(month_number, interest_paid, c='r', marker='.', markersize=10)
plt.plot(month_number, principal_paid, c='b', marker='.', markersize=10)
```

### MATLAB style vs. object oriented syntax
Below you will see example of object oriented syntax for Matplotlib

```
# example of tuple unpcacking
x, y = (3, 9)
```

```
# example of object oriented plotting syntax, compared to MATLAB 
fig, axes = plt.subplots(nrows=1, ncols=1)
axes.plot(month_number, interest_paid, c='b');
axes.plot(month_number, interest_paid, c='y');
```

```
# combination of both matlab and bject oriented syntax
fig, axes = plt.subplots(nrows=1, ncols=1)
plt.plot(month_number, interest_paid, c='k')
axes.plot(month_number, principal_paid, c='b')
```

### Setting Titles,  Labels and Limits for Matplotlib

```
# MATLAB-style setting xlim and ylim
plt.plot(month_number, interest_paid, c='r')
plt.plot(month_number, interest_paid, c='b')
plt.xlim(left=1, right=70)
plt.ylim(bottom=0, top=1000)
```

```
# setting labels for x and y
plt.plot(month_number, interest_paid, c='r')
plt.plot(month_number, interest_paid, c='b')
plt.xlabel('Month')
plt.ylabel('Dollars')
```

```
# setting title
plt.plot(month_number, interest_paid, c='r')
plt.plot(month_number, interest_paid, c='b')
plt.xlabel('Month')
plt.ylabel('Dollars')
plt.title('Interest and Principal Paid Each Month', fontsize=15)
```

```
# changing tick font size
plt.plot(month_number, interest_paid, c='r')
plt.plot(month_number, interest_paid, c='b')
plt.xlabel('Month')
plt.ylabel('Dollars')
plt.title('Interest and Principal Paid Each Month', fontsize=15)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
```

#### Setting Titles, Labels and Limits using Object-oreinted Matplotlib Syntax

```
fig, axes = plt.subplots(nrows=1, ncols=1)
axes.plot(month_number, interest_paid, c='g');
axes.plot(month_number, interest_paid, c='b');
```

```
# set xlim and ylim using object oriented syntax
fig, axes = plt.subplots(nrows=1, ncols=1)
axes.plot(month_number, interest_paid, c='g');
axes.plot(month_number, interest_paid, c='b');
axes.set_xlim(left=1, right=70)
axes.set_ylim(bottom=0, top=1000)
```

```
# set xlabel and ylabel object-oriented matplotlib syntax
fig, axes = plt.subplots(nrows=1, ncols=1)
axes.plot(month_number, interest_paid, c='g');
axes.plot(month_number, interest_paid, c='b');
axes.set_xlabel('Month')
axes.set_ylabel('Dollars')
```

```
# set title using object-oriented matplotlib syntax
fig, axes = plt.subplots(nrows=1, ncols=1)
axes.plot(month_number, interest_paid, c='g');
axes.plot(month_number, interest_paid, c='b');
axes.set_xlabel('Month')
axes.set_ylabel('Dollars')
axes.set_title('Interest and Principal Paid Each Month')
```

```
# changing fontsize using object-oriented matplotlib syntax
fig, axes = plt.subplots(nrows=1, ncols=1)
axes.plot(month_number, interest_paid, c='g');
axes.plot(month_number, interest_paid, c='b');
axes.set_xlabel('Month', fontsize=10)
axes.set_ylabel('Dollars', fontsize=10)
axes.set_title('Interest and Principal Paid Each Month', fontsize=15)
axes.tick_params(axis='x', labelsize=10)
axes.tick_params(axis='y', labelsize=10)
```

### Saving Plots to Files
